```python
# listenin elemanlarindan cift olanlarin karesini bir liste icinde dondurun while ile dongusunu yazin
sayilar_listesi = [1, 3, 4, 6, 5, 7]
x = 0
ciftlerin_kareler_listesi = []
while x < len(sayilar_listesi):
 if sayilar_listesi[x] % 2 == 0:
    ciftlerin_kareler_listesi.append(sayilar_listesi[x] ** 2)
 x += 1
print(ciftlerin_kareler_listesi)
```

```python
sayilar_listesi = [1,5,6,7,14,15,18,19,32,25]
y = 0
tek_elemanlar=[]
cift_elemanlar =[]

while y < len(sayilar_listesi):
    if sayilar_listesi[y] % 2 == 0:
      tek_elemanlar.append(sayilar_listesi[y] ** 2)
    else:
      cift_elemanlar.append(sayilar_listesi[y] * 2)
    y += 1
print(tek_elemanlar)
print(cift_elemanlar)
    
```
